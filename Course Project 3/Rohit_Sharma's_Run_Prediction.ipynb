## Aggregating ball by ball data to over by over data

```python
#creating over and balls from ball column which is in 0.1, 0.2 format for first over
deli["ball"] = deli["ball"].astype(str)
deli[["over", "ball_num"]] = deli["ball"].str.split(".", expand=True).astype(int)
deli["ball"] = deli["ball"].astype(float)
deli["over"] = deli["over"] + 1
deli["ball_left"] = 306 - (deli["over"]*6 + deli["ball_num"])
```

```python
#creating total runs column to calculate runs on each ball incusive of extras
deli["total_runs"] = deli["runs_off_bat"] + deli["extras"]
```

```python
#creating innings df so we can merger total runs for each innings with in original df
innings = deli.groupby(["match_id", "innings"])["total_runs"].sum().reset_index()
innings.rename(columns={"total_runs":"innings_total"}, inplace=True)
```

```python
innings
```

```python
#creating deliveries df and adding total runs
deliveries = deli.merge(innings, on=["match_id", "innings"])
deliveries
```

```python
#calculating cumulative runs
deliveries["cumulative_runs"] = deliveries.groupby(["match_id", "innings"])["total_runs"].cumsum()
```

```python
deliveries
```

## Creating new features like run rate, balls left, required run rate, wickets remaining

```python
#creating run rate column
deliveries["run_rate"] = deliveries.groupby(["match_id","innings"]).apply(lambda x: (x["cumulative_runs"]*6)/(300-x["ball_left"])).reset_index(level=[0,1], drop=True)
```

```python
deliveries
```

```python
#filtering for innings1 and 2 so we can add some columns respectively
innings_1 = deliveries[deliveries["innings"] == 1]

innings_1 = innings_1.groupby("match_id")["cumulative_runs"].max().reset_index()
innings_1.rename(columns={"cumulative_runs": "target"}, inplace=True)

innings_2 = deliveries[deliveries["innings"] == 2]
innings_2 = innings_2.merge(innings_1, on="match_id", how="left")
innings_2["target"] = innings_2["target"]+1
```

```python
innings_1
```

```python
innings_2
```

```python
#adding runs needed and rrr to innings 2
innings_2["runs_needed"] = innings_2["target"] - innings_2["cumulative_runs"]
innings_2["required_run_rate"] = (innings_2["runs_needed"]*6) / innings_2["ball_left"]
```

```python
#adding wickets remaining to innings 2
innings_2["wickets_remaining"] = innings_2.groupby("match_id")["wicket"].apply(lambda x: 10 - x.cumsum()).reset_index(level=0, drop=True)
```

```python
#innings 1 data
innings_1_f = deliveries[deliveries["innings"] == 1]
```

```python
#adding wickets remaining
innings_1_f["wickets_remaining"] = innings_1_f.groupby("match_id")["wicket"].apply(lambda x: 10 - x.cumsum()).reset_index(level=0, drop=True)
```

```python
#concatinating innings1,2
innings = pd.concat([innings_1_f,innings_2])
```

## Calculating Rohit Sharma's features based on historical data

```python
# predict rohit sharma's runs in the next match

# create column to store rohit sharma's runs till now
rohit["innings_score"] = rohit.groupby("match_id")["runs_off_bat"].cumsum()

# create column to store rohit sharma's balls faced till now
rohit["balls_faced"] = rohit.groupby("match_id").cumcount() + 1

# create column to store how many times rohit sharma got out till now
rohit["times_out"] = rohit["wicket"].cumsum()

# create column to store total runs of rohit sharma till now
rohit["total_rohit_runs"] = rohit["runs_off_bat"].cumsum()

# create column to store total balls faced by rohit sharma till now
rohit["total_rohit_balls"] = range(1, len(rohit) + 1)

# create column to store average of rohit sharma till now
rohit["average"] = rohit["total_rohit_runs"] / rohit["times_out"]

# create column to store rohit sharma's strike rate till now
rohit["strike_rate"] = (rohit["total_rohit_runs"]*100) / rohit["total_rohit_balls"]
rohit
```

## Splitting data into train and test

## Creating model and predicting Rohit Sharma's runs

```python
# train the classication model on the training data

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=100)
rf.fit(rohit_train_X, rohit_train_y)

y_pred = rf.predict(rohit_test_X)
print("Using Random Forest Classifier:")
print("Rohit's predicted runs: ", y_pred.sum())
print("Rohit's actual runs: ", rohit_test_y.sum())
```

```python
# xgboost for expected class as 0, 1, 2, 4, 6

from xgboost import XGBClassifier

# Define the custom label mapping
label_mapping = {0: 0, 1: 1, 2: 2, 4: 3, 6: 4}

# Map the labels using the custom mapping
y_train_mapped = [label_mapping[label] for label in rohit_train_y]
y_test_mapped = [label_mapping[label] for label in rohit_test_y]

xgb = XGBClassifier(n_estimators=100, max_depth=5, random_state=100)
xgb.fit(rohit_train_X, y_train_mapped)

y_pred_mapped = xgb.predict(rohit_test_X)

# Map the predicted labels to their original values
y_pred = [list(label_mapping.keys())[list(label_mapping.values()).index(label)] for label in y_pred_mapped]

print("Using XGBoost Classifier:")
print("Rohit's predicted runs: ", sum(y_pred))
print("Rohit's actual runs: ", rohit_test_y.sum())
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Assuming you have your data in X_train (features) and y_train (labels)
# Replace this with your actual data

# Encode class labels using LabelEncoder
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(rohit_train_y)
y_test = label_encoder.fit_transform(rohit_test_y)

# Build the DNN model
model = Sequential()
model.add(Dense(128, input_dim=rohit_train_X.shape[1], activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dense(len(label_encoder.classes_), activation='softmax'))  # Output layer with softmax activation for multi-class

# Compile the model
model.compile(loss='sparse_categorical_crossentropy',
              optimizer=Adam(learning_rate=0.001),
              metrics=['accuracy'])

# Train the model
model.fit(rohit_train_X, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=2)

# Make predictions on the test set
y_pred_probs = model.predict(rohit_test_X)
y_pred = [np.argmax(probabilities) for probabilities in y_pred_probs]

# Reverse the label encoding for evaluation
y_pred = label_encoder.inverse_transform(y_pred)

print("Using DNN:")
print("Rohit's predicted runs: ", y_pred.sum())
print("Rohit's actual runs: ", rohit_test_y.sum())
```
